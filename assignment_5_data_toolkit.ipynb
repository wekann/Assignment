```python
# Q1) Demonstrate three different methods for creating identical 2D arrays in Numpy .Provide the code for each method and the final output after each method.
# Three different methods for creating identical 2D arrays in NumPy:
# Method 1: Using np.zeros()
import numpy as np
array1 = np.zeros((3, 4))
array1.fill(5)
print("Method 1 Output:")
print(array1)

#Method 2: Using np.ones()
import numpy as np
array2 = np.ones((3, 4)) * 5
print("\nMethod 2 Output:")
print(array2)

#Method 3: Using np.full()
import numpy as np
array3 = np.full((3, 4), 5)
print("\nMethod 3 Output:")
print(array3)
```

```python
# Q3) Explain the following terms:
# The difference in np.array, np.asarray and np.asanyarray.
# The difference between Deep copy and shallow copy.
# Difference between `np.array`, `np.asarray`, and `np.asanyarray`:

#1.`np.array()` is a function that creates a new NumPy array from an array-like object or Python list.
# If the input is already a NumPy array, it will make a copy of that array unless the `copy` parameter is explicitly set to False.
# It always returns a new array, regardless of whether the input is already a NumPy array or not.

# 2.`np.asarray()` is similar to `np.array()` but with a few differences.
# It converts the input to an array, but if the input is already a NumPy array, it doesn't make a copy by default (unless the `copy` parameter is explicitly set to True).
# It aims to convert input to an array if it is not already an array-like object.

# 3.`np.asanyarray()` is also similar to `np.array()`, but it doesn't always create a new array.
# If the input is already a NumPy array, it returns the input array as it is without making a copy.
# It provides the flexibility to accept any array-like object or existing NumPy array as input.

### Difference between Deep copy and Shallow copy:

# 1. A shallow copy creates a new object but inserts references to the original objects into it.
# It means the copied object is not a completely new entity; rather, it contains references to the original nested objects.
# Changing elements of the copied object may affect the original object if they share references to nested objects.
# In Python, shallow copies can be made using the `copy()` method or using the `copy` module's `copy()` function.

# 2.A deep copy creates a new object and recursively inserts copies of the original objects into it.
# It means the copied object is entirely independent of the original object and its nested objects.
# Changing elements of the copied object will not affect the original object or its nested objects.
# In Python, deep copies can be made using the `copy` module's `deepcopy()` function.
```

```python
# Q14) Create a time-series dataset in a Pandas DataFrames with Column:"Data","Temperature","Humidity"and Perform the following tasks using Matpotlib:
# Plot the "temperature" and "humidity" on the same plot with different y-axes(left y-axes for "Temperature" and right y-axis for "Humidity").
# Label the x-axis as "Date".
# Set the title of the plot as "Temperature and Humidity Over Time".
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Date': pd.date_range(start='2024-01-01', periods=100),
    'Temperature': pd.Series(np.random.randn(100), name='Temperature'),
    'Humidity': pd.Series(np.random.randn(100), name='Humidity')
}
df = pd.DataFrame(data)
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
# Q15) Create a Numpy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matpotlib:
# Plot a histogram of the data with 30 bins.
# Overlays a line plot representing the normal distribution's probability density function(PDF).
# Label the x-axis as "Value" and the y-axis as "Frequency/Probability"
# Set the title of the plot as "Histogram with PDF Overlay"
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

```python
# Q17) Create a Seaborn Scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label te axes, and set the title as "Quadrant-wise Scatter Plot"
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(100)
y = np.random.randn(100)
quadrants = np.full_like(x, 'Quadrant I')
quadrants[(x < 0) & (y >= 0)] = 'Quadrant II'
quadrants[(x < 0) & (y < 0)] = 'Quadrant III'
quadrants[(x >= 0) & (y < 0)] = 'Quadrant IV'
data = {'X': x, 'Y': y, 'Quadrant': quadrants}
df = pd.DataFrame(data)
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis')
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

```python
# Q18) With Bokeh, plot a line chart of a sine wave function,add grid lines, label the axes, and set the title as "Sine Wave Function.  "
from bokeh.plotting import figure, output_file, show
import numpy as np
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5
output_file("sine_wave.html")
show(p)
```

```python
# Q19) Using Bokeh, generate a bar chart of randomly generated categorial data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as " Random Categorical Bar Chart
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool
import numpy as np
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))
p = figure(title="Random Categorical Bar Chart", x_axis_label='Categories', y_axis_label='Values')
hover = HoverTool()
hover.tooltips = [("Value", "@values")]
p.add_tools(hover)
colors = ['#FF5733', '#FFC300', '#DAF7A6', '#C70039', '#581845']
p.vbar(x=categories, top=values, width=0.5, color=colors)
output_file("random_categorical_bar_chart.html")
show(p)
```

```python
# Q20) Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as "Simple Line Plot"
import plotly.graph_objs as go
import numpy as np
x = np.linspace(0, 10, 100)
y = np.random.randn(100)
trace = go.Scatter(
    x=x,
    y=y,
    mode='lines',
    name='Random Data'
)
layout = go.Layout(
    title='Simple Line Plot',
    xaxis=dict(title='X-axis'),
    yaxis=dict(title='Y-axis')
)
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
# Q21) Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentage ,set the title as "Interactive Pie Chart"
import plotly.graph_objs as go
import numpy as np
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(labels))
trace = go.Pie(labels=labels, values=values, textinfo='label+percent')
layout = go.Layout(title='Interactive Pie Chart')
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
