```python
# Q5) Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
data = np.random.randn(100, 5)
cov_matrix = np.array([[1.0, 0.9, 0.3, 0.2, 0.1],
                       [0.9, 1.0, 0.2, 0.1, 0.3],
                       [0.3, 0.2, 1.0, 0.8, 0.5],
                       [0.2, 0.1, 0.8, 1.0, 0.7],
                       [0.1, 0.3, 0.5, 0.7, 1.0]])
data_correlated = np.dot(data, cov_matrix)
correlation_matrix = np.corrcoef(data_correlated, rowvar=False)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Q1)Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, output_file, show
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')
p.line(x, y, legend_label="sin(x)", line_width=2)
output_file("sine_wave.html")
show(p)
```

```python
# Q2) create a Bokeh scatter plot using randomly generated x and y value. Use different sizes and colors for the markers based on the "sizes" and "colors" columns.
from bokeh.plotting import figure, output_file, show
import numpy as np
import pandas as pd
np.random.seed(0)
data = {
    "x": np.random.rand(100) * 10,  # Generate random x values between 0 and 10
    "y": np.random.rand(100) * 10,  # Generate random y values between 0 and 10
    "sizes": np.random.rand(100) * 20 + 5,  # Generate random sizes between 5 and 25
    "colors": np.random.rand(100) * 100,  # Generate random colors between 0 and 100
}
df = pd.DataFrame(data)
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=df)
output_file("scatter_plot.html")
show(p)
```

```python
# Q3) Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ["Apples","Oranges","Bananas","Pears"]
# counts = [20,25,30,35]
from bokeh.plotting import figure, output_file, show
fruits = ["Apples", "Oranges", "Bananas", "Pears"]
counts = [20, 25, 30, 35]
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"
p.title.text_font_size = "16px"
output_file("bar_chart.html")
show(p)
```

```python
# Q4) Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist,bins = 30)
from bokeh.plotting import figure, output_file, show
import numpy as np
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram", background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
p.xaxis.axis_label = "Value"
p.yaxis.axis_label = "Frequency"
p.title.text_font_size = "16px"
output_file("histogram.html")
show(p)
```
